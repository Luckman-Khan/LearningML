```python
#red
shade = []
for i in range(256):
    t=[]
    for j in range(1920):
        t.append([i,0,0])
    shade.append(t)

for i in range(255,0,-1):
    t=[]
    for j in range(1920):
        t.append([i,0,0])
    shade.append(t)

#green
for i in range(256):
    t=[]
    for j in range(1920):
        t.append([0,i,0])
    shade.append(t)
for i in range(255,0,-1):
    t=[]
    for j in range(1920):
        t.append([0,i,0])
    shade.append(t)

#blue
for i in range(256):
    t=[]
    for j in range(1920):
        t.append([0,0,i])
    shade.append(t)

for i in range(255,0,-1):
    t=[]
    for j in range(1920):
        t.append([0,0,i])
    shade.append(t)

shade = np.array(shade)
plt.imshow(shade)
```

```python
cv.imwrite('shades.png',shade)
```

```python
import cv2 as cv
import matplotlib.pyplot as plt

# Load the image (as BGR)
bgr_img = cv.imread('retro-futuristic-space-travel.jpg')

if bgr_img is not None:
    # Convert from BGR to RGB
    rgb_img = cv.cvtColor(bgr_img, cv.COLOR_BGR2RGB)

    # Display side-by-side to see the difference
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.imshow(bgr_img)
    plt.title('Incorrect Colors (BGR)')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(rgb_img)
    plt.title('Correct Colors (RGB)')
    plt.axis('off')

    plt.show()
else:
    print("Error: Could not load the image.")
```
